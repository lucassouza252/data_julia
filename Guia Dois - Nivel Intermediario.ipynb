# Guia Intermediario de Julia

## Importação de Pacotes
<p>Importa os pacotes/bibliotecas que serão utilizados no código.</p>

```julia
# Pacote DataFrames é utilizado para manipular bases de dados(dataset).
# Pacote CSV é utilizado para importar arquivos do tipo CSV.
# Pacote StatsBase provê um apanhado de funções estatisticas.
# Pacote Plots é utilizado para plotar gráficos.
# Pacote StatPlots é um compilado de metodos estatísticos que trabalham em conjunto com Plots.
using DataFrames
using CSV
using StatsBase
using Plots
using StatPlots
```

## Leitura de arquivos
<p>Abre o arquivo que contem os dados que serão utilizados para análise.</p>

```julia
# Abre o arquivo CSV. O ponto-virgula serve para não mostrar a base de dados logo em seguida.
data = CSV.read("dataset.csv");
```

## Obter informações sobre a base de dados
<p>Funções que obtem certos tipos de informações que podem ser uteis para análisar a base de dados.</p>

```julia
# Identifica o "shape", tamanho, do dataset.
size(data)
```

```julia
# Mostra um numero n especifico dos primeiros dados do dataset, neste caso, 5.
head(data, 5)
```

```julia
# Descreve as caracteristicas do dado informado.
describe(data[:Peso])
```

```julia
# Realiza a contagem dos valores associado ao dado informado.
countmap(data[:Sexo])
```

## Visualização de Dados
<p>É a forma utilizada para ver o comportamento dos dados, ou para demonstrar algum tipo de informação por meio visual.</p>

```julia
# Graficos de dispersão são bons para ver a correlação entre as variáveis e como elas estão distribuidas.
Plots.scatter(data[:Idade], data[:Peso], labels="I x P", 
    title="Gráfico de Dispersão", xlabel="Idade", ylabel="Peso")
```

```julia
# Histograma demonstra a distribuição de frequência dos dados. 
Plots.histogram(data[:Idade], bins=10, xlabel="Idade", title="Histograma", labels="Idade")
```

```julia
# Grafico de Barras é bom para comparações, demonstrando os valores referentes a um dado.
Plots.bar(data[:Idade], data[:Filhos], xlabel="Idade", ylabel="Filhos", title="Grafico de Barras", labels="")
```
